```python
''' Problems Statement 1: Unique Event Participants
A college event has a list of students who registered multiple times by mistake
Write a program using set to store particiants Names and display only unique participants'''

Participants = [
  "Bob",
  "Bobby",
  "Mille",
  "Sadie",
  "Hopper",
  "Sadie",
  "Winona",
  "Bob"
]
unique_participants = set(Participants)
print(unique_participants)
```

```python
''' Problem Statement 2: Three  students choose elective subjects.
Find the common subjetcs chosen by all students using set operations.'''

electives_of_student1 = {"German","Finance","Psychology","Geography","Bee keeping"}
electives_of_student2 = {"German","Theatre","Finance","History","French","Engineering graphics"}
electives_of_student3 = {"German","Spanish","Geography","Political science","Engineering graphics","Finance"}

common_subjects = electives_of_student1 & electives_of_student2 & electives_of_student3
print("Their common subjects is/are:" ,common_subjects)

```

```python
''' Problem Statement 3: One set contains students in Cricket club and another
in football club. Find the students who are:
(1) Both clubs
(2) Only one club'''

Cricket_club = {"Kohli","Rohit","Glen Philips","Starc","David","Joe Root","Ronaldo"}
Football_club = {"Messi","Harry Kane","David","Benzama","Ronaldo","MBappe"}

both_clubs = Cricket_club & Football_club #Intersection
only_one_club = Cricket_club ^ Football_club # Symmetric difference

print("Students in both club: ",both_clubs)
print("Students in only one club: ",only_one_club)
```

```python
''' Problem Statement 4: Absent students
Given a set of all students and a set of present students, find the absent student'''

All_students = {"Rohit", "Ananya","Sara","Kriti","Ajay","Tanisha","Tara","Akash"}
Present_students = {"Ananya","Sara","Kriti","Ajay","Tanisha"}

Absent_students = All_students - Present_students
print("Absent students are: ",Absent_students)
```

```python
''' Problem Statement 5: Remove invalid entries.
A set contains valid course codes.
Remove a discontinued course code from the set.'''

course_codes = {"CS101","CS102","CS10","CS67","CS185"}
discontinued = "CS67"

course_codes.discard(discontinued)
print(course_codes)
```
