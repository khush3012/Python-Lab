```python
# 1(a) Check whether a number is positive or negative
# 1(b) Check whether a number is positive, negative or zero

number = int(input("Enter a number: "))

#1(a)
if number>=0:
  print("The number is positive")
else:
  print("The number is negative")

#1(b)
if number>0:
  print("The number is positive")
elif number==0:
  print("The number is zero")
else:
  print("the number is negative")
```

```python
# 8. Salary Calculation with allowances
''' Write a program to calculate gross salary of an employee based on basic salary
If basic<= 10K: HRA = 20%, DA = 80%
If basic<= 20K: HRA = 25%, DA = 90%
Otherwise HRA = 30%, DA = 95%'''

basic = float(input("Enter the salary:"))

if basic <= 10000:
  hra = basic * 0.20
  da = basic * 0.80
elif basic <=20000:
  hra = basic * 0.25
  da = basic * 0.90
else:
  hra = basic * 0.30
  da = basic * 0.95

gross_salary = basic + hra + da

print("Your net salary is: ", gross_salary)
```
