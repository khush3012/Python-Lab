#While Loop

```python
#While loop
#Used when number of iterations are not know
Syntax
while condition:
  statement(s)
```

```python
# Print numbers 1 to 5

i = 1
while (i<=5):
  print(i)
  i = i + 1
```

```python
# Take input from user, till what iteration the number will be printed? (print output in a single line)

n = int(input("Enter the number of iteratons: "))
i = 1
while i<=n:
  print(i, end = " ")
  i = i+1
```

```python
# Factorial of a number

n = int(input("Enter the number: "))
fact = 1
N = n
while n >= 1:
  fact = fact * n
  n = n - 1
print("Factorial is: ", fact)
```

```python
# Fibonacci Series using while loop

n = int(input("Enter number of terms: "))
a = 0
b = 1
i = 0
while i < n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    i = i + 1

```

```python
# Reverse a number

num = int(input("Enter the number: "))
rev = 0

while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10

print("Reversed number : ", rev)
```

```python
# Check Palindrome of a number or string

string = input("Enter a string: ")
rev = string[::-1]
if string == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Count digits in a number

num = int(input("Enter a number: "))
count = 0

while num > 0:
  count = count + 1
  num //= 10

print("Number of digits: ", count)
```

#While - Continue


```python
# Continue skips the remaining statements of the current iteration and jumps back to the while condition
```

```python
# Print numbers from 1 to 10 (skip 5)

i = 1
while(i<=10):
  if i == 5:
    i = i + 1
    continue
  print(i, end = " ")
  i = i + 1
```

```python
# Print only odd numbers between 1 to 10

i = 1
while (i<10):
  if i % 2 == 0:
    i = i + 1
    continue
  print(i, end = " ")
  i = i + 1

  # For even numbers we will write i % 2 != 0
```

```python
# Check whether the number is an Armstrong number or not

num = str(input("Enter a number "))
length = len(num)
sum = 0

for i in range(length):
  sum =sum + int(num[i])**length

if sum == int(num):
  print("Armstrong Number")

else:
  print("Not an Armstrong Number")

```

```python
# FLoyd's Traingle

n = int(input("Enter the number of rows: "))
number = 1

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(number, end=" ")
        number += 1
    print()
```
