```python
marks = [78,85,90,67,88]
print("Maximum mark:",max(marks))
print("Minimum marks",min(marks))
print("Sum of marks:",sum(marks))
print("Average of marks:",sum(marks)/len(marks))
print("Sorted marks:", marks.sort())
print(marks)
```

```python
'''Grocery Shopping List
A user maintains a grocery for monthly shopping'''

grocery_list = ["Milk","Eggs","Chips Ahoy cookies","Wheat pasta","Mushrooms","Cheese"]
print("Create a grocery list:" ,grocery_list)
grocery_list.append("Lemons")
print("Add new items:" ,grocery_list)
grocery_list.remove("Cheese")
print("Remove an item:" ,grocery_list)
print("Final shopping list:", grocery_list)
```

```python
# A person stores phone contacts in a list

contacts = ["Shanaya","Anushka","Preet","Kshitij","Ryan","Sydney","Teddy","Jenna"]
print("Create a contact list:", contacts)
contacts.append("Vishvam")
print("Add new contact:", contacts)
contacts.remove("Ryan")
print("Remove contact:", contacts)
contacts.sort()
print("Sorted contacts:", contacts)
```

```python
#Slicing Practice
list1=[20,30,50,60,40,10,70]
print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-2])
print(list1[4:6:-1])
print(list1[-2:-4:-2])
```

```python
# Temperature Analysis
temperature=[7,11,20,19,25,14,9,10,21,5]
print("Temprature of first five days :",temperature[0:5],"\nTemprature of last 5 days :",temperature[-5:])
print("Highest temperature :",max(temperature))
print("Lowest temperature :",min(temperature))
print("Average of the temperature :",sum(temperature)/len(temperature))
```
